# Visualizing Data using MatplotLib

A simple plot 

```python
from matplotlib import pyplot as plt 

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# Creates a line chart, years on x-axis, gdp on y-axis 
plt.plot(years, gdp, color = 'green', marker='o', linestyle='solid')

# Add a title
plt.title("Nominal GDP")

# add a label to the y-axis
plt.ylabel("Billions of $")
plt.show()
```

# Bar Chart 

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", " West Side Story"]

num_oscars = [5, 11, 3, 8, 10]

# plot bars with left x-coordinates [0,1,2,3,4], heights [num_oscars]

plt.bar(range(len(movies)), num_oscars)

plt.title(" My Fav Movie") # add a tittle
plt.ylabel(" # of Academic Awards") # Label the y-axis

# label x-axis with movie anmes at bar centers
plt.xticks(range(len(movies)),movies)

plt.show()
```

# Bar chart for plotting Histogram

```python
from collections import Counter

grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]

# Bucket grades by decile, but lets put 100 in the 90's 
histogram = Counter(min(grade // 10 * 10, 90)for grade in grades)

print(f' Histogram Values {histogram}')

plt.bar([x + 5 for x in histogram.keys()], # shift bars to the right by 5
        histogram.values(),                # Give each bar its correct Height
        10,                                # Give each bar a width of 10
        edgecolor = (0,0,0)                # Add a black boundary 
        )
plt.axis([-5,105,0,5])                     # X axis is from -5 to 105. Y axis is from 0 to 5

plt.xticks([10*i for i in range(11)])      # X labels from 0, 10, 20 etc
plt.yticks([i for i in range(6)])
plt.xlabel(" Decile")
plt.ylabel(" # of Students")
plt.title(" Destribution of Exam 1 Grades")
plt.show()
```

# Line Chart

```python
variance = [ 2**i for i in range(9)]
bias_squared = [ 2**i for i in range(9,0,-1)]
total_error = [x+y for x,y in zip(variance,bias_squared)]

xs = [i for i, _ in enumerate(variance)]

# we can make multiple calls to plt.plot
# to show multiple series on the same chart 
plt.plot(xs,variance, 'g-' , label = 'variance') # Green Solid line
plt.plot(xs,bias_squared, 'r-',label = 'Bias Squared') # red dot-dashed line
plt.plot(xs, total_error, 'b:' , label= 'Total Error') # Blue spotted line

# Because we assigned label to each line we can create a Legend. 
plt.legend(loc=9) # loc is the position of the Legend and 9 means top-Center
plt.xlabel("Model Complexity")
plt.xticks([])
plt.title(" The Bias Variance TradeOff")
plt.show()
```

# Scatterplots

```python
friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ["a", "b", "c", "d", "e", "f", "g", "h", "i"]

plt.scatter(friends, minutes)

# Label each point

for label , friend_count, minute_count in zip(labels,friends,minutes):
    plt.annotate(label,                             # Put the label with its point
                 xy=(friend_count,minute_count),     # Put the label with its point
                 xytext=(5, -5),                     # Adding a slight Offset
                 textcoords='offset points'
                 )
plt.title(" Daily minutes vs. Number of Friends")
plt.xlabel(" # of friends")
plt.ylabel(" Daily minutes spent on the site")
plt.show()
```
